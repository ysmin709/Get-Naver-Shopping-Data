```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('C:\\Users\\YSM\\Downloads\\result.csv')
df
```

```python
Rate = 0
```

```python
Jeju = df['Jeju']
Seoul = df['Seoul']
Daegu = df['Daegu']
Jeonju = df['Jeonju']
Gwangju = df['Gwangju']
Sejong = df['Sejong']
Changwon = df['Changwon']
AVG = df['AVG']
```

```python
max_Jeju = max(Jeju)
max_Seoul = max(Seoul)
max_Daegu = max(Daegu)
max_Jeonju = max(Jeonju)
max_Gwangju = max(Gwangju)
max_Sejong = max(Sejong)
max_Changwon = max(Changwon)
max_AVG = max(AVG)
```

```python
Rate_max_Jeju = 100
Rate_max_Seoul = round((max_Seoul / max_Jeju) * 100, 2)
Rate_max_Daegu = round((max_Daegu / max_Jeju) * 100, 2)
Rate_max_Jeonju = round((max_Jeonju / max_Jeju) * 100, 2)
Rate_max_Gwangju = round((max_Gwangju / max_Jeju) * 100, 2)
Rate_max_Sejong = round((max_Sejong / max_Jeju) * 100, 2)
Rate_max_Changwon = round((max_Changwon / max_Jeju) * 100, 2)
Rate_max_AVG = round((max_AVG / max_Jeju) * 100, 2)
```

```python
Rate_list = [Rate_max_Jeju, Rate_max_Seoul, Rate_max_Daegu, Rate_max_Jeonju, Rate_max_Gwangju, Rate_max_Sejong, Rate_max_Changwon, Rate_max_AVG]
```

```python
Rate_list
```

```python
max_df = pd.DataFrame(Rate_list, ['Jeju', 'Seoul', 'Daegu', 'Jeonju', 'Gwangju', 'Sejong', 'Changwon', 'AVG'], ['Rate of change'])
```

```python
max_df
```

```python
plt.title('Max Price Compared to Jeju')
# x축 라벨 설정
plt.xlabel('Area')
# x축 라벨 설정
plt.ylabel('Rate')

plt.ylim([40, 100])
# 선 그래프(x, y)
plt.plot(max_df.index, max_df['Rate of change'])
# 막대 그래프(x, y)
plt.bar(max_df.index, max_df['Rate of change'])
# 그래프 출력
plt.show()
```
