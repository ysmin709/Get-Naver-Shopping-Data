```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import matplotlib.ticker as ticker

plt.rcParams["axes.grid"] = True
plt.rcParams["font.size"] = 20
plt.rcParams["figure.figsize"] = (30,10)
```

```python
# Main Point In Graph - Line Graph 

plt.plot(df['Jeju'], color='red', marker='o', markersize=13, linestyle='--', linewidth=4.0)
plt.plot(df['Seoul'], color='magenta', marker='o', markersize=15, linestyle='--', linewidth=4.0)
plt.plot(df['Daegu'], color='black', marker='o', markersize=13, linestyle='--', linewidth=4.0)
plt.plot(df['Jeonju'], color='blue', marker='o', markersize=13, linestyle='--', linewidth=4.0)
plt.plot(df['Gwangju'], color='gold', marker='o', markersize=13, linestyle='--', linewidth=4.0)
plt.plot(df['Changwon'], color='lime', marker='o', markersize=13, linestyle='--', linewidth=4.0)

plt.legend(['Jeju', 'Seoul', 'Daegu', 'Jeonju', 'Gwangju', 'Changwon'], loc=1)
plt.title('Price Changes')
plt.xlabel("Month/Week")
plt.xticks(df.index[:], ['11/4 ', '11/5 ', '12/1 ', '12/2 ', '12/3 ', '12/4 ', '12/5 ', '1/1', '1/2', '1/3', '1/4', '1/5', '2/1', '2/2', '2/3', '2/4', '3/1'])
plt.ylabel("Price")
plt.ylim([1000, 4500])

plt.text(df.index[0], df.iloc[0][0], 'Jeju', fontsize=25, color='k', horizontalalignment='center', verticalalignment='top')
plt.text(df.index[0], df.iloc[0][2], 'Others', fontsize=25, color='k', horizontalalignment='center', verticalalignment='top')
plt.axvspan(df.index[3], df.index[14], facecolor='pink', alpha=0.3)

plt.show()
```

```python
# All Area's Price Before Price Changes Term - Line Graph

plt.plot(df['Jeju'], color='red', marker='o', markersize=13, linestyle='--', linewidth=4.0)
plt.plot(df['Seoul'], color='magenta', marker='o', markersize=15, linestyle='--', linewidth=4.0)
plt.plot(df['Daegu'], color='black', marker='o', markersize=13, linestyle='--', linewidth=4.0)
plt.plot(df['Jeonju'], color='blue', marker='o', markersize=13, linestyle='--', linewidth=4.0)
plt.plot(df['Gwangju'], color='gold', marker='o', markersize=13, linestyle='--', linewidth=4.0)
plt.plot(df['Changwon'], color='lime', marker='o', markersize=13, linestyle='--', linewidth=4.0)

plt.legend(['Jeju', 'Seoul', 'Daegu', 'Jeonju', 'Gwangju', 'Changwon'], loc=1)
plt.title('Price Changes')
plt.xlabel("Month/Week")
plt.xticks(df.index[:], ['11/4 ', '11/5 ', '12/1 ', '12/2 ', '12/3 ', '12/4 ', '12/5 ', '1/1', '1/2', '1/3', '1/4', '1/5', '2/1', '2/2', '2/3', '2/4', '3/1'])
plt.ylabel("Price")
plt.ylim([1000, 4500])

plt.text(df.index[0], df.iloc[0][0], 'Jeju', fontsize=25, color='k', horizontalalignment='center', verticalalignment='top')
plt.text(df.index[0], df.iloc[0][2], 'Others', fontsize=25, color='k', horizontalalignment='center', verticalalignment='top')
plt.axvspan(df.index[0], df.index[3], facecolor='lime', alpha=0.2)

plt.show()
```

```python
# All Area's Price Before Price Changes(Standard: Jeju) - Bar Graph

ax_y = []
for i in range(6):
    tmp = df.iloc[0][i] / 27
    ax_y.append(int(tmp))

plt.barh(df.index[1:6], ax_y[1:6], color=['dodgerblue', 'dodgerblue', 'dodgerblue', 'dodgerblue', 'dodgerblue'], alpha=0.7)
plt.yticks(df.index[1:6], ['Seoul', 'Daegu', 'Jeonju', 'Gwangju', 'Changwon'], fontsize=15, rotation='0')
plt.xlabel("%", fontsize='20')
plt.ylabel("Area", fontsize='20')
plt.xlim([0, 100])
for i in range(1, 6):
    plt.text(ax_y[i], df.index[i], str(ax_y[i]) + '%', fontsize=25, color='k', horizontalalignment='left', verticalalignment='top')
```

```python
# All Area's Max price for Price Changes Term - Bar Graph

max_list = []
for i in range(6):
    tmp = 0
    for j in range(len(df)):
        tmp = max(tmp, df.iloc[j][i])
    max_list.append(int(tmp))

plt.bar(df.index[:6], max_list[:6], color=['dodgerblue', 'dodgerblue', 'dodgerblue', 'dodgerblue', 'dodgerblue', 'dodgerblue'], alpha=0.7)
plt.xticks(df.index[:6], ['Jeju', 'Seoul', 'Daegu', 'Jeonju', 'Gwangju', 'Changwon'], fontsize=15, rotation='0')
plt.xlabel("Area", fontsize='20')
plt.ylabel("Price", fontsize='20')
for i in range(0, 6):
    plt.text(df.index[i], max_list[i], max_list[i], fontsize=25, color='k', horizontalalignment='center', verticalalignment='bottom')
```
